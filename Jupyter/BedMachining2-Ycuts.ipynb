```python
import re
import math

def arrayToFile(path, data):
    with open(path, "w") as txt_file:
        for line in data:
            txt_file.write(" ".join(line))
```

```python
#Input Values
xLegnth = 1125
yLegnth = 500

xMargin = 20 + 21
yMargin = 20 + 21

xspacing=21
yspacing=21

#Bed Fram 0,0
#X=-545.48
#Y=326.60
#Z=126.17
X=0
Y=0
Z=0

z1=Z-1.5
z2=Z+50

#CaulcationOfProbePoints
xStartPos = X+xMargin
xEndPos = X+xLegnth-xMargin

yStartPos = Y+yMargin
yEndPos = Y+yLegnth-yMargin

xTravel=abs(xStartPos)+abs(xEndPos)
yTravel=abs(yStartPos)+abs(yEndPos)

xProbePointNumber=xTravel/xspacing
yProbePointNumber=yTravel/yspacing

x=xStartPos
y=yStartPos-yspacing


#IteratingThroughPoints
array = ['0'] * (int(2*yProbePointNumber)+8)
i = 0

#Setting the start
newString = "G1 X549.325 Y354.544 Z130.155\n"
array[i]=[newString]
i += 1


#setting start position to the side of bed
X = format((x),".3f")
Z2 = format((z2),".3f")
Y = format((y),".3f")

Xon=format((x-(xMargin+30)),".3f")
Xoff=format((x+(xMargin+30)),".3f")



newString = "G1 X{} Y{} Z{} \n".format(Xon, Y, Z2)
array[i]=[newString]
i += 1
newString = "G1 X{} Y{} Z{} \n".format(Xon, Y, Z2)
array[i]=[newString]
i += 1

#turning on the spindle
newString = "M03\n"
array[i]=[newString]
i += 1

a = True
while (y<yEndPos):
    if a is True:
        x=xStartPos
        X = format((x),".3f")
        Z2 = format((z2),".3f")
        Y = format((y),".3f")
        Z1 = format((z1),".3f")
        newString = "G1 X{} Y{} Z{} \n".format(X, Y, Z1)
        array[i]=[newString]
        i+=1
        x=xEndPos
        X = format((x),".3f")
        newString = "G1 X{} Y{} Z{} \n".format(X, Y, Z1)
        array[i]=[newString]
        i+=1
    if a is False:
        x=xEndPos
        X = format((x),".3f")
        Y = format((y),".3f")
        Z1 = format((z1),".3f")
        Z2 = format((z2),".3f")
        newString = "G1 X{} Y{} Z{} \n".format(X, Y, Z1)
        array[i]=[newString]
        i+=1
        x=xStartPos
        X = format((x),".3f")
        newString = "G1 X{} Y{} Z{} \n".format(X, Y, Z1)
        array[i]=[newString]
        i+=1
    a = not a
    y+=(yspacing)
X = format((x),".3f")
Z2 = format((z2),".3f")
Y = format((y),".3f")


#Setting return point
newString = "G1 X{} Y{} Z{} \n".format(Xoff, Y, Z1)
array[i]=[newString]
i += 1
#turning off the spindle
newString = "M05\n"
array[i]=[newString]
i += 1

#Setting return point
newString = "G1 X{} Y{} Z{} \n".format(Xoff, Y, Z2)
array[i]=[newString]
i += 1

newString = "G1 X549.325 Y354.544 Z130.155\n"
array[i]=[newString]
i += 1
arrayToFile("SpindleFinal7.gcode", array)
```
